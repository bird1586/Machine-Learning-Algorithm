# 主成分分析-最大化變異數觀點(Principal Component Analysis – Maximum Variance Prospective)

##### 先引入我們需要的packages

```python
import os 
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

# Training Data

```python
N=3000
CLASS=2
d=6
X=np.zeros((N,d))
base=random.normal(loc=1, scale=10, size=N)
for i in range(0,d):
    X[:,i]=12*np.sin(base)**(i+1)+random.rand(N)
```

# Plot Distribition

```python
plt.rcParams["figure.figsize"] = (18,18)
fig , ax = plt.subplots()
fig.subplots_adjust(hspace=0.1, wspace=0.1)
for i in range(1,d+1):
    for j in range(1,d+1):
        plt.subplot(d, d, (i-1)*d+j)
        plt.plot(X[:,i-1],X[:,j-1],"ko",markersize=1)
        plt.text(0.5, 0.5, str((i, j)), fontsize=18, ha='center')
plt.show()
```

# Maximum Variance Prospective

```python
T=6
S=np.sum(X[:,:,None]*X[:,None,:],axis=0)/N
W,V=np.linalg.eig(S)
B=V[:,:T]
X_fit=np.dot(np.dot(B,B.T),X.T).T
plt.rcParams["figure.figsize"] = (18,18)
fig , ax = plt.subplots()
fig.subplots_adjust(hspace=0.1, wspace=0.1)
for i in range(1,d+1):
    for j in range(1,d+1):
        plt.subplot(d, d, (i-1)*d+j)
        plt.plot(X[:,i-1],X[:,j-1],"o",markersize=1)
        plt.plot(X_fit[:,i-1],X_fit[:,j-1],"o",markersize=1)
        plt.text(0.5, 0.5, str((i, j)), fontsize=18, ha='center')
plt.show()
```
